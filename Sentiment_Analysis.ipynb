```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
```

```python
print("\n... INDUVIDUAL CALLS [POSITIVE FOLLOWED BY NEGATIVE] ...\n")
print('\t', classifier("I have been waiting for a HUGGINGDACE Course my whole life"))
print('\t', classifier("I hate this so much"))
```

```python
print("\n\n... LIST CALL [POSITIVE FOLLWED BY NEGATIVE] ...\n")
print('\t', classifier(["I have been waiting for a HUGGINGDACE Course my whole life",
                        "I hate this so much",
                        ]))
```

```python
def pretty_zero_shot(_classifier, _sequences, _candidate_labels):
  for k,v in classifier(
        sequences=_sequences,
        candidate_labels=_candidate_labels
  ).items(): print(f"\t{k}:\n\t\t--> {v}")
```

```python
example_info={
      "EXAMPLE 1":{
            "sequences":"This session for learning about the transformer library",
            "candidate_labels":["education", "politics","business",]
      },
      "EXAMPLE 2":{
            "sequences":"I dont have to work today since its a weekend",
            "candidate_labels":["monday", "tuesday","wednesday","thursday", "friday","saturday","sunday"],
      },
      "EXAMPLE 3":{
            "sequences":"I hate when people burst crackers due to the noise, what is so special about today",
            "candidate_labels":["monday", "tuesday","wednesday","thursday", "friday","saturday","sunday"],
      },
      "EXAMPLE 4":{
            "sequences":"the lincoln material",
            "candidate_labels":["person", "place","thing"],
      },

      "EXAMPLE 5":{
            "sequences":"2+2=",
            "candidate_labels":["1", "2","3","4","5"],
      },
      "EXAMPLE 6":{
            "sequences":"Hello, today, I'm going to talk about the majority of humans that own and operate call phones on a daily basis",
            "candidate_labels":["lead", "position","claim","counterclaim", "evidence","concluding statement"],
      },
      "EXAMPLE 7":{
            "sequences":"Chips and icecream",
            "candidate_labels":["healthy snack", "unhealthy snack"],
      },
      "EXAMPLE 8":{
            "sequences":"Celery and Cake",
            "candidate_labels":["healthy snack", "unhealthy snack"],
      },
}
```

```python
#instantiting a zero shot classifier
classifier=pipeline("zero-shot-classification")
```

```python
for ex_name, ex_map in example_info.items():
    print(f"\n\n\n{ex_name}: ")
    pretty_zero_shot(classifier, ex_map["sequences"], ex_map["candidate_labels"])
```

```python
#instantiate the default text generation model
generator=pipeline("text-generation")
```

```python
print("\n\n\n\m... EXAMPLE#1...\n")
original_text="In this course, i will teach you how to"
pred_text=generator(original_text, return_full_text=False, num_return_sequences=1)
print(f"\n==== Original Phrase ====\n\t--> {original_text} ...")
print(f"\n==== Generated Text ====\n\t--> {pred_text[0]['generated_text']} ...")
```
